### **Задание 6. (бонус, 1 балл)**

Реализуйте в классе DecisionTree поддержку параметров max_depth, min_samples_split и min_samples_leaf по аналогии с DecisionTreeClassifier. Постройте графики зависимости качества предсказания в зависимости от этих параметров для набора данных tic-tac-toe (см. следующее задание).

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from hw4_code_pro import DecisionTree

url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/tic-tac-toe/tic-tac-toe.data'
data = pd.read_csv(url, header=None)

for col in data.columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col].astype(str))

X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
feature_types = ['categorical'] * X.shape[1]

def score_for(**params):
    tree = DecisionTree(feature_types=feature_types, **params)
    tree.fit(X_train, y_train)
    pred = tree.predict(X_val)
    return accuracy_score(y_val, pred)

depths = list(range(1, 11))
scores_depth = [score_for(max_depth=d) for d in depths]

splits = [2, 5, 10, 20, 50]
scores_split = [score_for(min_samples_split=s) for s in splits]

leafs = [1, 2, 5, 10, 20]
scores_leaf = [score_for(min_samples_leaf=l) for l in leafs]

plt.figure(figsize=(12, 3))
plt.subplot(1, 3, 1)
plt.plot(depths, scores_depth, marker='o')
plt.title('max_depth')
plt.xlabel('value')
plt.ylabel('accuracy')

plt.subplot(1, 3, 2)
plt.plot(splits, scores_split, marker='o')
plt.title('min_samples_split')
plt.xlabel('value')
plt.ylabel('accuracy')

plt.subplot(1, 3, 3)
plt.plot(leafs, scores_leaf, marker='o')
plt.title('min_samples_leaf')
plt.xlabel('value')
plt.ylabel('accuracy')

plt.tight_layout()
````

**1. Загрузите следующие наборы данных** (напомним, что pandas умеет загружать файлы по url, в нашем случае это файл \*.data), предварительно ознакомившись с описанием признаков и целевой переменной в каждом из них (они записаны в Data Folder, в файле *.names):

* [nursery](https://archive.ics.uci.edu/ml/datasets/Nursery) (классы записаны в последнем столбце, считаем, что not_recom и recommend — класс 0, very_recom, priority, spec_prior — класс 1).

**2. Закодируйте категориальные признаки, использовав LabelEncoder.** С помощью cross_val_score (cv=10) оцените accuracy на каждом из этих наборов данных следующих алгоритмов:
* DecisionTree, считающий все признаки вещественными

* DecisionTree, считающий все признаки вещественными + one-hot-encoding всех признаков
* DecisionTreeClassifier из sklearn. Запишите результат в pd.DataFrame (по строкам — наборы данных, по столбцам — алгоритмы).

**Рекомендации:**
* Чтобы cross_val_score вычисляла точность, нужно передать scoring=make_scorer(accuracy_score), обе фукнции из sklearn.metrics.
* Если вам позволяет память (а она скорее всего позволяет), указывайте параметр sparse=False в OneHotEncoder (если вы, конечно, используете его). Иначе вам придется добиваться того, чтобы ваша реализация дерева умела работать с разреженными матрицами.

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, make_scorer
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.tree import DecisionTreeClassifier
from hw4_code_pro import DecisionTree

scorer = make_scorer(accuracy_score)

def load_data():
    data = {}

    mush = pd.read_csv('agaricus-lepiota.data', header=None)
    data['mushrooms'] = (mush.iloc[:, 1:], mush.iloc[:, 0])

    ttt = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/tic-tac-toe/tic-tac-toe.data', header=None)
    data['tic-tac-toe'] = (ttt.iloc[:, :-1], ttt.iloc[:, -1])

    cars = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data', header=None)
    y_cars = cars.iloc[:, -1].replace({'unacc': 0, 'acc': 0, 'good': 1, 'vgood': 1})
    data['cars'] = (cars.iloc[:, :-1], y_cars)

    nursery = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/nursery/nursery.data', header=None)
    y_nur = nursery.iloc[:, -1].replace({
        'not_recom': 0, 'recommend': 0, 'very_recom': 1, 'priority': 1, 'spec_prior': 1
    })
    data['nursery'] = (nursery.iloc[:, :-1], y_nur)

    return data

def label_encode_df(df):
    df = df.copy()
    for c in df.columns:
        le = LabelEncoder()
        df[c] = le.fit_transform(df[c].astype(str))
    return df

class MyTree(BaseEstimator, ClassifierMixin):
    def __init__(self, feature_types, one_hot=False):
        self.feature_types = feature_types
        self.one_hot = one_hot
        self._encoder = None
        self._model = None

    def fit(self, X, y):
        X = np.asarray(X)
        y = np.asarray(y)
        if self.one_hot:
            self._encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
            X = self._encoder.fit_transform(X)
            feature_types = ['real'] * X.shape[1]
        else:
            feature_types = self.feature_types
        self._model = DecisionTree(feature_types=feature_types)
        self._model.fit(X, y)
        return self

    def predict(self, X):
        X = np.asarray(X)
        if self.one_hot:
            X = self._encoder.transform(X)
        return self._model.predict(X)

data = load_data()
rows = []

for name, (X_df, y) in data.items():
    X_enc = label_encode_df(X_df)
    X = X_enc.values
    y = np.asarray(y)

    ft_real = ['real'] * X.shape[1]
    ft_cat = ['categorical'] * X.shape[1]

    model_real = MyTree(feature_types=ft_real, one_hot=False)
    model_cat = MyTree(feature_types=ft_cat, one_hot=False)
    model_oh = MyTree(feature_types=ft_real, one_hot=True)
    sk_model = DecisionTreeClassifier(random_state=42)

    scores_real = cross_val_score(model_real, X, y, cv=10, scoring=scorer)
    scores_cat = cross_val_score(model_cat, X, y, cv=10, scoring=scorer)
    scores_oh = cross_val_score(model_oh, X, y, cv=10, scoring=scorer)
    scores_sk = cross_val_score(sk_model, X, y, cv=10, scoring=scorer)

    rows.append({
        'dataset': name,
        'DT real': scores_real.mean(),
        'DT categorical': scores_cat.mean(),
        'DT one-hot': scores_oh.mean(),
        'sklearn DT': scores_sk.mean()
    })

result = pd.DataFrame(rows).set_index('dataset')
result
````
